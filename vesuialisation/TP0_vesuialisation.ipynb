# TP0 — Étape 2 : Charger et explorer les données


```python
# Importer les bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

## Q1 — Charger le fichier iris.csv

```python
df = pd.read_csv('Iris.csv')
df.head()
```

## Q2 — Afficher les premières et dernières lignes

```python
print(df.head())
print(df.tail())
```

## Q3 — Info sur les colonnes et types

```python
df.info()
```

## Q4 — Espèces présentes et effectifs

```python
df['Species'].value_counts()
```

## Q5 — Colonnes numériques seulement

```python
num_df = df.select_dtypes(include=np.number)
num_df.head()
```

## Q6 — Répartition des espèces

```python
df['Species'].value_counts().plot(kind='bar')
plt.show()
```

## Q7 — Pairplot

```python
sns.pairplot(df, hue='Species')
```

## Q8 — Visualisation 3D

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.iloc[:,0], df.iloc[:,1], df.iloc[:,2])
plt.show()
```

## Q9 — Corrélations

```python
corr = num_df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

## Q10 — ACP (PCA) en 3D

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(num_df)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2])
plt.show()
```
